```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

def rk4_step(func, y, t, dt, m):
    k1 = dt * func(y, t, m)
    k2 = dt * func(y + k1/2, t + dt/2, m)
    k3 = dt * func(y + k2/2, t + dt/2, m)
    k4 = dt * func(y + k3,   t + dt, m)
    y += (1/6) * (k1 + 2*k2 +2*k3 + k4)
    return y

def rk4(func, init, t, m):
    num_variables = len(init)
    y = np.zeros((len(t), num_variables))
    y[0,:] = init
    for i in range(0, len(t)-1):
        dt = t[i+1] - t[i]
        y[i+1,:] = rk4_step(func, y[i,:], t[i], dt, m)
    return y
```

```python
# Global constants for earth
G = 6.6742*10**-11 # gravitational constant, Nm^2/kg^2
M = 5.9722*10**24 # mass of earth, kg
R = 6371*10**3 # radius of earth, m
omega = 7.2921e-5

# Global constants for rocket
m = 299999999999
A = np.pi*8.7**2
C_d = 0.1

m_shuttle = 80e3
m_primary = 1.2e6
m_secondary = 100e3

t_burn_primary = 127

dm_dt1 = m_primary / t_burn_primary
dm_dt2 = 1000

F_T_primary = 36e6 # This is the default primary thurst, but you can input your own values if you wish!
F_T_secondary = 5e6 # This value should not really be changed!

# Initial conditions
v0 = 7.29e-5*R # tangential velocity at earth's surface = omega * R
init1 = np.array([0, R, v0, 0])
```

```python
def force_gravity(m, x, y):
    '''Calculate the force of gravity on a mass m due to earth.
    
    Parameters:
    
    m : mass of object [kg]
    x : x distance away from the center of earth [m]
    y : y distance away from the center of earth [m]
    
    Returns:
    
    F_g_x, F_g_y: Force on mass m due to gravity in x and y directions respectively [N]'''
    
    # force magnitude
    F_g = G*M*m/(x**2 + y**2)
    
    # Polar angle measured counterclockwise from the x axis
    theta = np.arctan2(y,x)
    
    # Draw a simple free-body diagram to derive these. The gravitational force is always radially inward.
    # Break the vector into components and you can easily see the relationshp coded below. I'll upload a diagram soon.
    F_g_x = F_g * np.cos(theta)
    F_g_y = F_g * np.sin(theta)
    
    return F_g_x, F_g_y

def air_density(m, x, y):
    '''Calculate (a close approximation of) the air density at a given distance away from the center of earth.
    
    Parameters:
    
    m : mass of object [kg]
    x : x distance away from the center of earth [m]
    y : y distance away from the center of earth [m]
    
    Returns:
    
    Density at point (x,y) [kg/m^3]'''
    
    
    rho_0 = 1.225 # sea level density, kg/m^3
    L = 0.0065 # temperature lapse rate, K/m
    T = 288.15 # temperature, K. Temperature is obviously NOT constant and this is a simplification
    M = 0.0289644 # molar mass of dry air, kg/mol
    R_gas = 8.31447 # ideal gas constant, J/(mol K)
    
    F_g_x, F_g_y = force_gravity(m, x, y) 
    g = np.sqrt(F_g_x**2 + F_g_y**2)/m # acceleration due to gravity at (x, y) [m/s^2]
    r = np.sqrt(x**2 + y**2)
    
    # Equation pulled from Wikipedia
    rho = rho_0 * np.exp(-g*M*(r-R)/(R_gas*T))
    
    # @TODO: Consider atmosphere...

    return rho

def force_drag(m, x, y, vx, vy):
    '''Return the drag force and object experiences during flight.
    
    Parameters:
    
    m : mass of object [kg]
    x : x distance away from the center of earth [m]
    y : y distance away from the center of earth [m]
    vx : x velocity of object
    vy : y velocity of object
    
    Returns:
    
    F_drag_x, F_drag_y: Force due to drag in x and y directions respectively [N]'''
    
    rho = air_density(m, x, y)
    
    b = 1/(2*m) * rho * A * C_d
    
    # Dr. Wood-Vasey derived these in class for hw7 and hw8. The formulas are identical to his work.
    F_drag_x = b*np.sqrt(vx**2 + vy**2)*vx
    F_drag_y = b*np.sqrt(vx**2 + vy**2)*vy

    return F_drag_x, F_drag_y


def acceleration(m, t, x, y, vx, vy):
    '''Return the net acceleration of the rocket.
    
    Parameters:
    
    m : mass of object [kg]
    t : time [s]
    x : x distance away from the center of earth [m]
    y : y distance away from the center of earth [m]
    vx : x velocity of object
    vy : y velocity of object
    
    Returns:
    
    acceleration_net_x, acceleration_net_y : net acceleration in the x and y directions respectively [m/s^2]
    '''
    
    # Retrieve mass
    m = mass(t)
    
    # Retrieve forces 
    F_g_x, F_g_y = force_gravity(m, x, y)
    F_drag_x, F_drag_y = force_drag(m, x, y, vx, vy)
    F_T_x, F_T_y = force_thrust(t, x, y)
    
    # Gravitational forces are always attractive and drag forces oppose the direction of motion.
    # Hence, both of those are negative. Thrust is obviously positive.
    acceleration_net_x = (-F_g_x - F_drag_x + F_T_x)/m
    acceleration_net_y = (-F_g_y - F_drag_y + F_T_y)/m

    return acceleration_net_x, acceleration_net_y

def F(init, t, m):
    '''Function to be integrated.
    
    Parameters:
    
    init : array of initial conditions [x, y, vx, vy]
    t : time
    m : mass of rocket at time t
    
    Returns:
    
    array [vx, vy, ax, ay]'''

    x, y = init[0], init[1]
    vx, vy = init[2], init[3]
    ax, ay = acceleration(m, t, x, y, vx, vy)
    
    return np.array([vx, vy, ax, ay])


def print_init(init):
    '''Print initial conditions
    
    Parameters:
    
    init : array of initial conditions [x, y, vx, vy]
    
    Returns:
    
    None
    Prints out all relevant constants and initial conditions'''
    
    print('INITIAL CONDITIONS')
    print('x =', init[0])
    print('y =', init[1])
    print('vx =', init[2])
    print('vy =', init[3])
    
    print('\nSPECS')
    print('F_T_primary', F_T_primary)
    print('F_T_secondary', F_T_secondary)
    print('m_shuttle =', m_shuttle)
    print('m_primary =', m_primary)
    print('m_secondary', m_secondary)
    print('dm_dt_primary', dm_dt1)
    print('dm_dt_secondary', dm_dt2)
```

This is where the process gets a little convoluted. The purpose of this cell is to determine when we should start
the secondary boosters and how long we should keep them burning.

General idea:

We get enough radial height with our primary boosters, but we do not get
tangential velocity--the rocket eventually falls back to earth. So, the 
purpose of the secondary boosters is solely to provide more tangential 
velocity.

So, we need to figure out when we should start them and how long they
should burn.

If we run the simulation without burning the secondary booster fuel, we can
find the time at which the rocket reaches its maximum radial height, i.e.
when it's radial velocity is zero.

Here's the plan. We want to reach our orbital velocity right at that time.

Step 1: Calculate the maximum radial height and find the time at which it
occurs.

Step 2: Calculate the tangential velocity at this time and the tangential
velocity that we need in order to maintain an orbit at this radial height.
The difference between these two values is the velocity boost we need in
order to get into orbit (typically around 2000 m/s)

Step 3: Now that we know how much more tangential velocity we need, we can
explicitly calculate how long the boosters need to burn in order to reach
this value:

$m_0 = m_{shuttle} + m_{secondary}$ 

Note that the mass from the primary fuel is not considered, because it was
previously burned.

$r_{secondary} = 1000 kg/s$

Rate at which secondary fuel burns.

$$a(t) = \frac {F_{secondary}}{m_0 - r_{secondary} \cdot t}$$

Integrating $a(t)$ will give us $\Delta v_{secondary}$:

$$\Delta v_{secondary} = \int_0^t \frac {F_{secondary}}{m_0 - r_{secondary} \cdot t} dt = \Big[-\frac{F_{secondary}}{r_{secondary}} \cdot \ln(m - r_{secondary}\Big]_0^t$$

$$\Delta v_{secondary} = -\frac{F_{secondary}}{r_{secondary}} \cdot \ln\Big(\frac{m - r_{secondary} \cdot t}{m}\Big)$$

Solve for $t$ yields:

$$ t = \frac{m_0}{r_{secondary}} \cdot \Big[1 - \exp \Big(\frac{r_{secondary} \cdot \Delta v_{secondary}} {F_{secondary}}\Big)\Big]$$

This $t = t_{secondary}$ tells us how long the secondary boosters should burn!!!!!

Because we want to reach orbital velocity right when we achieve our maximum radial height, we must start the secondary boosters $t_{secondary}$ seconds before we reach out maximum radial height.

This process assure that the rocket achieves a nearly circular orbit. It's not perfect, but it is certainly very close.

IMPORTANT NOTE:

The built in odeint function is used to run this simulation, but our own rk4 function was used in the final simulation.
This is because odeint does not use equal time steps. If we were to use odeint to run the final simulation, there is a
very good chance that the entire secondary boost stage would be skipped! Hence, we must use an integrate method that uses equal time steps. The rk4 function takes a bit longer to integrate, but we must deal with this, because odeint flat out does not give us the correct result.

```python
F_T_primary = input('What force should the primary thrusters supply? ' 
                    'Reasonable values should be no less than 25e6 N, '
                    'or else the rocket\n won\'t get enough radial height '
                    'and inevitably crash into the earth AND CRASH THE PROGRAM!!!!!: ')
F_T_primary = float(F_T_primary)

def mass(t):
    '''Returns mass of rocket at time t.
    
    Parameters:
    
    t : time
    
    Returns:
    
    mass of rocket at t
    
    Notes:
    
    This mass function does not burn the secondary booster fuel. It is merely used in order to determine when
    to start and stop secondary boosters. So it's not the "official" mass function.
    '''
    
    if t <= t_burn_primary:
        
        return m_shuttle + m_secondary +  m_primary - dm_dt1 * t
        
    return m_shuttle + m_secondary

def force_thrust(t, x, y):

    if t <= t_burn_primary:
        
        return (((F_T_primary) * np.sin(np.pi/2 * t/t_burn_primary)), 
                 (F_T_primary) * np.cos(np.pi/2 * t/t_burn_primary))
    
    return 0,0

def approximate_boost_info(init, t_max, N):
    '''Calculate when to start and stop secondary boosters.
    
    Parameters:
    
    init : array containing initial conditions [x, y, vx, vy]
    t_max : how long the launch simulation should run
    N : number of data points
    
    Returns:
    
    time to start secondary boosters, time to stop secondary boosters, time interval for which the secondary booster are burning'''
    
    # Create time array
    T = np.linspace(0, t_max, N)
    
    # Integrate F
    data = odeint(F, init, T, (m,))

    # Extract data
    x, y, vx, vy = data[:,0], data[:, 1], data[:,2], data[:,3]
    
    # Calculate necessary polar quantities
    r = np.sqrt(x**2+y**2)
    theta = np.arctan2(y, x)
    vr = vx*np.cos(theta) + vy*np.sin(theta)
    vtheta = - vx*np.sin(theta) + vy*np.cos(theta)
    
    # Find max radial height and the time at which it occurs
    max_r_index = np.argmax(r)
    max_r_time = T[max_r_index]
    max_r = np.max(r)
    
    # Find tangential velocity at our maximum radial height
    max_r_vtheta = vtheta[max_r_index]
    
    # Calculate how much more velocity we need to get into orbit
    v_boost = np.sqrt(G*M/max_r) - abs(max_r_vtheta)
    
    # Calculate how long the boosters should burn
    t_boost = (m_shuttle + m_secondary)/dm_dt2 * (1 - np.exp(-dm_dt2 * v_boost / F_T_secondary))
    
    print('\nMax height of', max_r, 'm at t =', max_r_time, 's')
    print('vtheta =', max_r_vtheta, 'm/s at this time.\n' 
          'We need a boost of', v_boost, 'm/s to get to orbital speed at this height.')
    print('Fire secondary boosters at t =', max_r_time-t_boost, 's for', t_boost, 's')
    print('Run the next cell to run the final simulation!')
    
    return max_r_time-t_boost, max_r_time, t_boost


t_secondary_start, t_secondary_stop, t_burn_secondary = approximate_boost_info(init1,10000,10000)
```

```python
def mass(t):
    '''Returns mass of rocket at time t.
    
    Parameters:
    
    t : time
    
    Returns:
    
    mass of rocket at t
    
    Notes:
    
    This mass function DOES burn the secondary booster fuel.  It IS the "official" mass function.
    '''
        
    if t <= t_burn_primary:
        
        return m_shuttle + m_secondary + m_primary - dm_dt1 * t
    
    elif t > t_burn_primary and t < t_secondary_start:
        
        return m_shuttle + m_secondary
    
    elif t >= t_secondary_start and t <= t_secondary_stop:

        return m_shuttle + m_secondary - dm_dt2 * (t - t_secondary_start)

    return m_shuttle + m_secondary - dm_dt2 * (t_burn_secondary)


def force_thrust(t, x, y):
    theta = np.arctan2(y, x)
    
    if t <= t_burn_primary:
        
        return (((F_T_primary) * np.sin(np.pi/2 * t/t_burn_primary)), 
                 (F_T_primary) * np.cos(np.pi/2 * t/t_burn_primary))
    
    elif t >= t_secondary_start and t <= t_secondary_stop:

        return F_T_secondary * np.sin(theta), - F_T_secondary * np.cos(theta)
    
    return 0,0



def test(init, t_max, N):
    
    print_init(init)
    
    # Create time array
    T = np.linspace(0, t_max, N)
        
    # Integrate F
    data = rk4(F, init, T, m)
    
    # Extract data
    x, y, vx, vy = data[:,0], data[:, 1], data[:,2], data[:,3]
        
    plt.figure(figsize=(20,3))
    
    # Plot trajectory
    plt.subplot(1,3,1)
    plt.plot(x, y)
    plt.title('Trajectory')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.axis('equal')
    plt.plot(0, 0, 'ro')

    # Draw a circle to represent earth
    earth = plt.Circle((0,0), R, color='green')
    plt.subplot(1,3,1).add_patch(earth)
        
    # Plot vx vs t
    plt.subplot(1,3,2)
    plt.plot(T, vx)
    plt.title('vx vs t')
    plt.xlabel('t (s)')
    plt.ylabel('vx (m/s)')

    # Plot vy vs t
    plt.subplot(1,3,3)
    plt.plot(T, vy)
    plt.title('vy vs t')
    plt.xlabel('t (s)')
    plt.ylabel('vy (m/s)')
    plt.show()
    
    plt.figure(figsize=(20,3))
        
    # Plot r vs t
    plt.subplot(1,3,1)
    plt.plot(T, np.sqrt(x**2+y**2))
    plt.title('r vs t')
    plt.xlabel('t (s)')
    plt.ylabel('r (m)')
    
    theta = np.arctan2(y, x)
    vtheta = - vx*np.sin(theta) + vy*np.cos(theta)
    vr = vx*np.cos(theta) + vy*np.sin(theta)
    
    plt.subplot(1,3,2)
    plt.plot(T, vtheta)
    plt.title('vtheta vs t')
    plt.xlabel('t (s)')
    plt.ylabel('vtheta (m/s)')
    
    plt.subplot(1,3,3)
    plt.plot(T, vr)
    plt.title('vr vs t')
    plt.xlabel('t (s)')
    plt.ylabel('vr (m/s)')
    plt.show()
    
    # Create 3D plot
    fig = plt.figure(figsize=(20,20))
    ax = fig.gca(projection='3d')
    
    # Set z = 0 for all times t
    z = np.zeros_like(x)

    # Plot trajectory in 3D
    ax.plot(x, y, z, 'm', linewidth='4')
    
    # Plot initial and final positions
    ax.plot([x[0]], [y[0]], [z[0]], 'bo', label='inital position')
    ax.plot([x[-1]], [y[-1]], [z[-1]], 'co', label='final position')
    
    plt.legend(loc='best')
    
    # Create wireframe sphere to represent earth
    u, v = np.mgrid[0:2*np.pi:30j, 0:np.pi:30j]
    X = R*np.cos(u)*np.sin(v)
    Y = R*np.sin(u)*np.sin(v)
    Z = R*np.cos(v)
    ax.plot_wireframe(X, Y, Z, color="green")
    
    # Set axes limits. We want the scales to be equal, and we want to make sure we include the entire trajectory.
    # So, set the maximum limit to the largest x, y, or z value that appears 
    max_limit = max(np.max(x), np.max(y), np.max(z))
    ax.set_xlim(-max_limit, max_limit)
    ax.set_ylim(-max_limit, max_limit)
    ax.set_zlim(-max_limit, max_limit)
    
    plt.title('3D Trajectory Plot')

    plt.show()

t_max = 40000
test(init1, t_max, t_max)
```
