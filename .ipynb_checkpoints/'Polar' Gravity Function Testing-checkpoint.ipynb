```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from numpy.testing import assert_almost_equal

# Global constants
m = 1
A = 1
C_d = 1
```

```python
def force_gravity(m, x, y):
    '''Calculate the force of gravity on an object.
    
    Parameters:
    
    m : mass of object [kg]
    y : distance above surface of earth [m]
    
    returns:
    
    F_g : force due to gravity [N]'''
    
    G = 6.6742*10**-11 # gravitational constant, Nm^2/kg^2
    M = 5.9722*10**24 # mass of earth, kg
    R = 6371*10**3 # radius of earth, m
    
    F_g = G*M*m/(x**2 + (y)**2)
    return F_g

def print_init(init):
    print('INITIAL CONDITIONS')
    print('x =', init[0])
    print('y =', init[1])
    print('vx =', init[2])
    print('vy =', init[3])
```

```python
def x_acceleration1(x, y):
    theta = np.arctan2(y,x)
    F_g = -force_gravity(m, x, y)/m
    return F_g * np.cos(theta)

def y_acceleration1(x, y):
    cos_theta = y/(x**2 + y**2)**(1/2)
    F_g = -force_gravity(m, x, y)/m
    return F_g * cos_theta

def F1(init, T):
    x, y = init[0], init[1]
    vx, vy = init[2], init[3]
    ax, ay = x_acceleration1(x, y), y_acceleration1(x, y)
    
    return np.array([vx, vy, ax, ay])

def test_F1(init, t_max, N):
    T = np.linspace(0, t_max, N)
    
    data = odeint(F1, init, T)
    x, y, vx, vy = data[:,0], data[:, 1], data[:,2], data[:,3]
    
    plt.figure(figsize=(20,4))
    
    plt.subplot(1,3,1)
    plt.plot(x, y)
    plt.title('Trajectory')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.axis('equal')
    plt.plot(0, 0, 'ro')

    plt.subplot(1,3,2)
    plt.plot(T, x)
    plt.title('x vs t')
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')

    plt.subplot(1,3,3)
    plt.plot(T, y)
    plt.title('y vs t')
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')
    plt.show()
    
def plot_test_F1():
    G = 6.6742*10**-11 # gravitational constant, Nm^2/kg^2
    M = 5.9722*10**24 # mass of earth, kg
    R = 6371*10**3 # radius of earth, m
    
    y = R
    v0 = np.sqrt(G*M)/(0 + (y)**2)**(1/4)
    
    init1 = np.array([0, y, v0, 0])

    print_init(init1)
    test_F1(init1, 10000, 10000)

    #print_init(init2)
    #test_F1(init2, t_ground2*1.15, 100000)

    #print_init(init3)
    #test_F1(init3, t_ground3*1.15, 100000)

    #print_init(init4)
    #test_F1(init4, t_ground4*1.15, 100000)
    
    G = 6.6742*10**-11 # gravitational constant, Nm^2/kg^2
    M = 5.9722*10**24 # mass of earth, kg
    R = 6371*10**3 # radius of earth, mplot_test_F1()
```

```python
G = 6.6742*10**-11 # gravitational constant, Nm^2/kg^2
M = 5.9722*10**24 # mass of earth, kg
R = 6371*10**3 # radius of earth, m


def air_density(x, y):
    '''Calculate (a close approximation of) the air density at a given 
    height.
    
    Parameters:
    
    y : height above sea level [m]
    
    Returns:
    
    rho : density at height h [kg/m^3]'''
    
    
    rho_0 = 1.225 # sea level density, kg/m^3
    L = 0.0065 # temperature lapse rate, K/m
    # Temperature is obviously NOT constant and this is a simplification
    T = 288.15 # temperature, K
    M = 0.0289644 # molar mass of dry air, kg/mol
    R_inf = 8.31447 # ideal gas constant, J/(mol K)
    g = force_gravity(m, x, y)/m # acceleration due to gravity at height h [m/s^2]
    
    r = np.sqrt(x**2 + y**2)
    
    rho = rho_0 * np.exp(-g*M*(r-R)/(R_inf*T))
    return rho



def force_drag(x, y, vx, vy, variable_density=True):
    '''Return the drag force the object experiences during flight.
    
    Parameters:
    
    v : velocity [m/s]
    h : height [m]
    
    Returns:
    
    F_drag: force due to drag [N]'''
    
    rho = air_density(x, y)
    
    # for testing purposes
    if not variable_density:
        rho = 1.16
    
    b = 1/(2*m) * rho * A * C_d
        
    F_drag_x = b*np.sqrt(vx**2 + vy**2)*vx
    F_drag_y = b*np.sqrt(vx**2 + vy**2)*vy
    #print(x, y, rho)
    return F_drag_x, F_drag_y
```

```python
def x_acceleration4(x, y, vx, vy):
    theta = np.arctan2(y,x)
    F_g = -force_gravity(m, x, y)
    F_g_x = F_g * np.cos(theta)
    F_drag_x, F_drag_y = force_drag(x, y, vx, vy)
    #return F_g_x/m
    return (F_g_x - F_drag_x)/m

def y_acceleration4(x, y, vx, vy):
    theta = np.arctan2(y,x)
    F_g = -force_gravity(m, x, y)
    F_g_y = F_g * np.sin(theta)
    F_drag_x, F_drag_y = force_drag(x, y, vx, vy)
    #return F_g_y/m
    return (F_g_y - F_drag_y)/m

def F4(init, T):
    x, y = init[0], init[1]
    vx, vy = init[2], init[3]
    ax, ay = x_acceleration4(x, y, vx, vy), y_acceleration4(x, y, vx, vy)
    
    return np.array([vx, vy, ax, ay])




def test_F4(init, t_max, N):
    T = np.linspace(0, t_max, N)
    
    data = odeint(F4, init, T)
    x, y, vx, vy = data[:,0], data[:, 1], data[:,2], data[:,3]
    
    print(x[-1], y[-1], np.sqrt(x[-1]**2 + y[-1]**2))
    
    plt.figure(figsize=(20,3))
    
    plt.subplot(1,3,1)
    plt.plot(x, y)
    plt.title('Trajectory')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.axis('equal')
    plt.plot(0, 0, 'ro')
    
    earth = plt.Circle((0,0), R, color='green')
    plt.subplot(1,3,1).add_patch(earth)

    plt.subplot(1,3,2)
    plt.plot(T, x)
    plt.title('x vs t')
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')

    plt.subplot(1,3,3)
    plt.plot(T, y)
    plt.title('y vs t')
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')
    plt.show()
    
    plt.figure(figsize=(20,3))
    
    plt.subplot(1,2,1)
    plt.plot(T, vx)
    plt.title('vx vs t')
    plt.xlabel('t (s)')
    plt.ylabel('vx (m/s)')
    
    plt.subplot(1,2,2)
    plt.plot(T, vy)
    plt.title('vy vs t')
    plt.xlabel('t (s)')
    plt.ylabel('vy (m/s)')
    
    plt.show()
    
def plot_test_F4():
    G = 6.6742*10**-11 # gravitational constant, Nm^2/kg^2
    M = 5.9722*10**24 # mass of earth, kg
    R = 6371*10**3 # radius of earth, m
    
    y = R + 250e3
    v0 = np.sqrt(G*M)/(0 + (y)**2)**(1/4)
    
    init1 = np.array([0, y, v0, 0])
    init2 = np.array([0, R*3, 3310, 1000])
    init3 = np.array([0, R*2, 7310, 1000])
    init4 = np.array([0, R + 300e3, v0, 0])
    init5 = np.array([0, R+150e3, v0*0.8, v0*0.8])
    
    print_init(init1)
    test_F4(init1, 25*10000, 100000)
    
    print_init(init2)
    test_F4(init2, 100000, 100000)
    
    print_init(init3)
    test_F4(init3, 1000000, 100000)
    
    print_init(init4)
    test_F4(init4, 10000, 100000)
    
    print_init(init5)
    test_F4(init5, 10000, 100000)
    
    
plot_test_F4()

# Launching from the surface of the earth is not exactly possible at this point, because we NEED a thrusting force!
# The resistive drag force is too much to get the rocket into orbit. If we change the initial conditions to make the
# rocket launch at a ridiculous speed, even that won't work, because the drag force increases with an increasing
# velocity...So, all of these tests are done at initial heights decently far above the surface of the earth.
```
